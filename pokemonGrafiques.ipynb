```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Type 1', data=df, palette='bright')
plt.title('Bar plot de pokémon por tipo')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.show()
```

```python
plt.stackplot(df['Generation'], df['HP'], df['Attack'], df['Defense'], df['Sp. Atk'], df['Sp. Def'], df['Speed'],
              labels=['HP', 'Attack', 'Defense', 'Sp. Attack', 'Sp. Defense', 'Speed'])

plt.xlabel('Generación')
plt.ylabel('Estadística')
plt.title('Stack plot de estadisticas por generación')
plt.legend(loc='upper left')

plt.show()
```

```python
gen1 = df[df['Generation'] == 1]
plt.figure(figsize=(10, 6))
plt.stem(gen1['Total'], linefmt='C0-', markerfmt='o', basefmt=' ')
plt.xlabel('Estadísticas totales')
plt.ylabel('Número de Pokémon')
plt.title('Stem Plot de distribución de las estadísticas totales de los Pokémon de la primera generación')
plt.grid(True)
plt.show()
```

```python
gen3 = df[df['Generation'] == 3]

plt.figure(figsize=(10, 6))
plt.boxplot(gen3['Attack'])
plt.xlabel('Ataque')
plt.ylabel('Valor')
plt.title('Box plot distribución de los ataque en la tercera generación')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'HP', 'Speed']])
plt.title('Box plot distribución de estadísticas de los Pokémon')
plt.ylabel('Valor')
plt.xlabel('Estadística')
plt.show()
```

```python
tipo_counts = df['Type 2'].value_counts()

plt.pie(tipo_counts, labels=tipo_counts.index, autopct='%1.1f%%')
plt.title('Pie plot de distribución de tipos secundarios de Pokémon')
plt.show()
```
