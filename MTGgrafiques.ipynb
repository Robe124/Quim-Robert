```python
import pandas as pd
import numpy as np
from PIL import Image
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('MTG-OTJ-draft-card-ratings-2024-05-25_2.0.csv')
```

```python
# Exercici 1: crea un histograma per veure la cantitat de cartes que pertanyen al mateix color
```

```python
color_counts = df['Color'].value_counts()
plt.figure(figsize=(9, 6))
color_counts.plot(kind='bar', color='skyblue')
plt.title('Cantitat de cartes per color')
plt.xlabel('Color')
plt.ylabel('Número de cartes')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# Exercici 3: crea un histograma de colors per veure la cantitat de cartes segons la seva raresa 
```

```python
tipo_carta_counts = df['Rarity'].value_counts()
colors = np.random.rand(len(color_counts), 3) # Aquí generem colors aleatoris

plt.figure(figsize=(9, 5))
tipo_carta_counts.plot(kind='bar', color=colors) # I aquí assignem els colors a les barres
plt.title('Cantitat de cartes per raresa')
plt.xlabel('Raresa')
plt.ylabel('Cantitat')
plt.xticks(rotation=0)
plt.grid(axis='y')
```

```python
# Exercici 4: crea un box plot segons el winrate de les cartes Comunes en GIH (Games in Hand Win) 
```

```python
df['GIH WR'] = df['GIH WR'].str.replace('%', '').astype(float) # Treiem el símbol '%' perque dona problemes
comunes = df[df['Rarity'] == 'C']

plt.figure(figsize=(9, 5))
plt.boxplot(comunes['GIH WR'])
plt.xlabel('Winrate')
plt.ylabel('Valor')
plt.title('Box plot de winrate GIH en cartas comunes')
plt.show()
```

```python
# Exercici 5: fes una gráfica pie plot per represantar quin percentatge de cartes tenen un GIH WR major a 55%, 
# quines tenen entre 55% i 50%, i quines tenen menys de 50%.
```

```python
menos_50 = comunes[comunes['GIH WR'] < 50].shape[0]
entre_50_55 = comunes[(comunes['GIH WR'] >= 50) & (comunes['GIH WR'] <= 55)].shape[0]
mas_55 = comunes[comunes['GIH WR'] > 55].shape[0]

labels = ['Menys de 50% WR', 'Entre 50% i 55% WR', 'Més de 55% WR']
sizes = [menos_50, entre_50_55, mas_55]
colors = ['#ff9999','#66b3ff','#99ff99']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribució de GIH WR')
plt.axis('equal')  # Para asegurar que el gráfico sea circular
plt.show()
```
