```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('MTG-OTJ-draft-card-ratings-2024-05-25_2.0')
```

```python
# Ejercicio 1: Crear un histograma para ver la cantidad de cartass segun el color de dicha carta 
```

```python
#kdfngnjdfg
# Paso 2: Supongamos que la columna que contiene el color se llama 'Color'
# Agrupar por color y contar cuántas cartas hay de cada uno
color_counts = df['Color'].value_counts()

# Paso 3: Crear el histograma
plt.figure(figsize=(10, 6))
color_counts.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Cartas por Color')
plt.xlabel('Color')
plt.ylabel('Número de Cartas')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
